```python
import pandas as pd
import json
import re
from tqdm import tqdm, trange, tnrange

import os
import sys
```

```python
DataDir = 'Data/'
```

```python
os.listdir(DataDir)
```

```python
dflist = []
for i in os.listdir(DataDir):
    print()
    if 'csv' in i:
        print(i, '\n', '-'*30)
        df = pd.read_csv(DataDir+'//'+i)
        print(df.columns)
        print(df.head())
        print('-'*60)
        dflist.append(df)
    if 'json' in i:
        print(i, '\n', '-'*30)
        df = pd.read_json(DataDir+'//'+i)
        print(df.columns)
        df = df[['id', 'body']]
        df.columns=['ID', 'Joke']
        print(df.head())
        print()
        dflist.append(df)
```

```python
df = pd.concat(dflist)
```

```python
df.reset_index(inplace=True)
df = df[['Joke']]
```

```python
df
```

```python
df.to_csv(DataDir+'dataall.txt', header=None, index=None, sep=' ', mode='a')
```
